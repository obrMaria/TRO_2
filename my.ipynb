```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

обозначим жесткость пружины, массу груза и работу пружины, совершаемую в ходе выполнения опыта

каждой характеристике присвоим значение

```python
k = 2.5
m = 0.3
A = 0.8
f = 0
```

по формуле найдем частоту колебаний 
$$ \omega = \sqrt{\frac{k}{m}} $$

```python

omega = math.sqrt(k / m)
print(omega)
```

```python
t = np.linspace(0, 10, 100)
print(t)
```

маятник совершает гамронические колбания по закону $ x = A *\sin(\omega*t + \phi) $

```python
x = A * np.sin(omega*t + f)
plt.plot(t,x)
```

В результате выполнения следующего кода переменная F будет содержать массив значений силы сопротивления воздуха для каждой точки времени t

Для расчета силы сопротивления воздуха в маятнике существует формула:
$$F = -0.5 * rho * S * Cd * v^2$$

где:
>F - сила сопротивления воздуха  
>rho - плотность воздуха  
>S - площадь поперечного сечения маятника  
>Cd - коэффициент лобового сопротивления  
>v - скорость маятника  

Задаим дополнительные параметры маятника

```python
rho = 1.2  
S = 0.1   
Cd = 0.5  
```

```python
v = np.sin(t) # просто пример скорости
F = -0.5 * rho * S * Cd * v**2
print(F)
```

```python
plt.plot(t,F)
```
