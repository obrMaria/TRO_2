```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 2.5
m = 0.3
A = 0.8
phio = 0
```

```python
omega = math.sqrt(k / m)
print(omega)
```

```python
t = np.linspace(0,7,111)
```

```python
print(t)
```

```python
x = A * np.sin(phio + omega*t)
plt.plot(t,x)
```
