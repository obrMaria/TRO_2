```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

обозначим жесткость пружины, массу груза и работу пружины, совершаемую в ходе выполнения опыта

каждой характеристике присвоим значение

```python
k = 2.5
m = 0.3
A = 0.8
f = 0
```

по формуле найдем частоту колебаний 
$$ \omega = \sqrt{\frac{k}{m}} $$

```python
omega = math.sqrt(k / m)
print(omega)
```

```python
t = np.linspace(0,7,100)
print(t)
```

маятник совершает гамронические колбания по закону $$ x = A *\sin(\omega*t + \phi) $$ 

```python
x = A * np.sin(omega*t + f)
plt.plot(t,x)
```
